```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import urllib
from bs4 import BeautifulSoup
import datetime
import ast
import csv 
from IPython.display import HTML
%matplotlib inline
```

```python
URL_1 = 'https://www.indeed.ca/jobs?q=data+scientist&l=Canada&start'
main_page = urllib.request.urlopen(URL_1)
content = main_page.read()

soup_1 = BeautifulSoup(content, 'lxml')
next_pages_links_prep = []
for divTag in soup_1.find_all('div', {'class': 'pagination'}):
    for aTag in divTag.find_all('a'):
        next_pages_links_prep.append(aTag.get('href'))

# Remove the last element in next_pages_links_prep which is simillar to the first element in the list
next_pages_links = next_pages_links_prep[:-1]

# Get the common portion among all next_pages_links' elements into a list of strings 
# ..then join the elements to form a single string with the common portion of the URL using the .join() method
common_portion_in_next_pages_links_prep = list(next_pages_links[len(next_pages_links)-1])[:38]
common_portion_in_next_pages_links = ''.join(common_portion_in_next_pages_links_prep)

# Create an array of 34 'common_portion_in_next_pages_links' using 
array_of_common_portion_in_next_pages_links = []
for iu in np.arange(0, 34):
    array_of_common_portion_in_next_pages_links.append(common_portion_in_next_pages_links)
    
# Create a list of numbers from 200 to 860 covering all pages yielded from the search Job Title "Data Scientist" in "Canada" 
# ..and convert it to a list of string elements
additional_pageNumbers_list_prep1 = [str(d) for d in str(np.arange(200,880,20))]

# strip the list to remove '[' and ']' and ' '
additional_pageNumbers_list_prep2 =[]
for stg in additional_pageNumbers_list_prep1:
    if (stg != '[') & (stg != ']') & (stg != ' ') & (stg != '\n'): 
          additional_pageNumbers_list_prep2.append(stg)

# Combine each 3 digits strings into a single element by switching the list to a numpy array then applying the reshape() method 
# ..then apply the tolist() method to switch the numpy array back to a list object (list of lists)
additional_pageNumbers_list_prep3 = np.array(additional_pageNumbers_list_prep2).reshape(34,3).tolist()

# then join the elements of each 3-digit list to form a single string with the string of 3-digit number using the .join() method
additional_pageNumbers_list =[]
for item in additional_pageNumbers_list_prep3:
    concatenated = ''.join(item)
    additional_pageNumbers_list.append(concatenated)

# Switch the two lists of lists to numpy arrays for element wise addition (concatination)
numpyArray_array_of_common_portion_in_next_pages_links = np.array(array_of_common_portion_in_next_pages_links)
numpyArray_additional_pageNumbers_list = np.array(additional_pageNumbers_list)

# Element wise concatination using .core.defchararray.add() method
numpyArray_of_URLs = np.core.defchararray.add(numpyArray_array_of_common_portion_in_next_pages_links, numpyArray_additional_pageNumbers_list)

List_of_additional_URLs_prep1 = numpyArray_of_URLs.tolist()

# Add the links for next pages available on the first page to the list of all pages' links on all next pages to obtain the URLs for all pages 
List_of_all_URLs_prep1 = next_pages_links + List_of_additional_URLs_prep1

# Add the site portion to the  'List_of_all_URLs_prep1' and add the 1st page's URL 
# ..to obtain complete list of URLs to feed into urllib.request.urlopen() method in a for List_of_additional_URLs_prep1# 
site_portion = 'https://www.indeed.ca'
List_of_all_URLs = []
for link in List_of_all_URLs_prep1:
    link = site_portion + link
    List_of_all_URLs.append(link)
List_of_all_URLs.insert(0, URL_1)
List_of_all_URLs
```

```python
job_data = []
macro_skills_dict = {}
def find_job_info(job_search_results):
    for x in job_search_results:
        job_title = x.find('a',{'data-tn-element': "jobTitle"}).text.strip()
        job_link = "https://www.indeed.ca" + x.find('a')['href']
        job_company = x.find('span',{'class': "company"}).text.strip()
        try:
            job_salary = x.find('span',{'class': "salary no-wrap"}).text.strip()
        except AttributeError:
            job_salary = "Salary Not found"
        try:
            job_location = x.find('div',{'class': "location"}).text.strip()
        except AttributeError:
            job_location = x.find('span',{'class': "location"}).text.strip()
        job_data.append([job_title, job_link, job_company, job_salary, job_location])
        # extract skills from detail page
        temp_url = urllib.request.urlopen(job_link)
        temp_content = temp_url.read()
        temp_soup = BeautifulSoup(temp_content, 'lxml')
        # turn html into text
        souptxt = temp_soup.text
        # initialize indicators and search for skills
        skills_dict = {
            'Excel' : 0,
            'Python': 0,
            'R': 0,
            'Java': 0,
            'C/C++': 0,
            'MATLAB': 0,
            'SAS': 0,
            'SQL': 0,
            'SPSS': 0,
            'Tableau': 0,
            'Hadoop': 0,
            'Spark': 0,
            'communication': 0,
            'presentation': 0,
            'problem solving': 0,
            'project management': 0,
            'consulting': 0,
            'leadership': 0
            }
        if souptxt.find('Excel') != -1:
            skills_dict['Excel'] = 1
        if souptxt.find('Java') != -1:
            skills_dict['Java'] = 1 
        if souptxt.find('Python') != -1:
            skills_dict['Python'] = 1  
        if souptxt.find('R,') != -1:
            skills_dict['R'] = 1
        if souptxt.find('C/C++') != -1:
            skills_dict['C/C++'] = 1
        if souptxt.find('MATLAB') != -1:
            skills_dict['MATLAB'] = 1
        if souptxt.find('SAS') != -1:
            skills_dict['SAS'] = 1
        if souptxt.find('SQL') != -1:
            skills_dict['SQL'] = 1
        if souptxt.find('SPSS') != -1:
            skills_dict['SPSS'] = 1
        if souptxt.find('Tableau') != -1:
            skills_dict['Tableau'] = 1
        if souptxt.find('Hadoop') != -1:
            skills_dict['Hadoop'] = 1
        if souptxt.find('Spark') != -1:
            skills_dict['Spark'] = 1
        if souptxt.find('communication') != -1:
            skills_dict['communication'] = 1    
        if souptxt.find('presentation') != -1:
            skills_dict['presentation'] = 1
        if souptxt.find('problem solving') != -1:
            skills_dict['problem solving'] = 1
        if souptxt.find('project management') != -1:
            skills_dict['project management'] = 1
        if souptxt.find('consulting') != -1:
            skills_dict['consulting'] = 1
        if souptxt.find('leadership') != -1:
            skills_dict['leadership'] = 1
        macro_skills_dict[job_link] = skills_dict

for page in List_of_all_URLs:
    page = urllib.request.urlopen(page)
    content = page.read()
    soup = BeautifulSoup(content, 'lxml')
    results = soup.find_all('div',{'class': 'jobsearch-SerpJobCard row result'})
    find_job_info(results)

    results = soup.find_all('div',{'class': 'jobsearch-SerpJobCard row sjlast result'})
    find_job_info(results)

    results = soup.find_all('div',{'class': 'jobsearch-SerpJobCard lastRow row result'})
    find_job_info(results)

    for data in job_data:    
        print(data[0])
        print(data[1])
        print(data[2])
        print(data[3])
        print(data[4]) 
        print(macro_skills_dict[data[1]])
        print("\n******************\n")
```
