#### Loop over all job links on results page

```python
results_dict = {} # empty dict to store results
for job_link in soup.find_all('a', {'data-tn-element':'jobTitle'}):
    job_title = job_link.get('title')
    results_dict[job_title] = {}
    results_dict[job_title]['job_id'] = job_link.get('id')
results_df = pd.DataFrame(results_dict).T
results_df
```
